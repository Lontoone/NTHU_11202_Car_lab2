```python
import numpy as np        
import matplotlib.pyplot as plt
size = (600,525)
center = (100,200)

x = np.abs( np.arange(size[1]) - center[1])
y = np.abs( np.arange(size[0]) - center[0])
print(x.shape)
#xx,yy = np.meshgrid(x, y)

z = np.array(np.meshgrid(x, y)).T

print(z.shape)
print(z)
plt.imshow(z[:,:,0]+z[:,:,1])
plt.show()
```
